```python
import astropy
import numpy as np
import pandas as pd
import matplotlib
import scipy

from astropy.io import fits
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['text.color'] = 'Black'
plt.rcParams['font.size'] = '14'
plt.rcParams['figure.figsize'] = (11, 7)
```

```python
def IR_excess(file1, file2, file3, title, vmin = 0, vmax = 0, y = 0.9):
    if vmax == 0 & vmin == 0:
        plt.imshow(file2 / (file1 + file3), cmap = 'magma')
    elif vmax != 0:
        plt.imshow(file2 / (file1 + file3), cmap = 'magma', vmax = vmax)
    elif vmin != 0 & vmax == 0:
        plt.imshow(file2 / (file1 + file3), cmap = 'magma', vmin = vmin)
    else:
        plt.imshow(file2 / (file1 + file3), cmap = 'magma', vmin = vmin, vmax = vmax)
    plt.colorbar()
    plt.title(title, fontsize = 30)
```

```python
file11 = fits.open('/Users/FRANNYPACK/Downloads/grind!/DRAO/Data/Fortier_NHI/1H0717/_inter_HI.fits')
file12 = fits.open('/Users/FRANNYPACK/Downloads/grind!/DRAO/Data/Fortier_NHI/1H0717/_IR.fits')
file13 = fits.open('/Users/FRANNYPACK/Downloads/grind!/DRAO/Data/Fortier_NHI/1H0717/_local_HI.fits')
```

```python
IR_excess(file11[0].data, file12[0].data, file13[0].data, '1H0717', vmax = 1.2)
```
